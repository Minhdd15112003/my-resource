Số chính phương là số tự nhiên có thể biểu diễn dưới dạng bình phương của một số tự nhiên khác.
Ví dụ: 
- 0 = 0^2
- 1 = 1^2  
- 4 = 2^2
- 9 = 3^2
- 16 = 4^2
- 25 = 5^2
- 36 = 6^2
...

Để xác định một số có phải số chính phương hay không, ta lấy căn bậc 2 của số đó. Nếu kết quả là số nguyên thì số ban đầu là số chính phương.
